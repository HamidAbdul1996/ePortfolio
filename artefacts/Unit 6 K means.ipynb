```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 

%matplotlib inline
```

```python
import pandas as pd
df = pd.read_csv('C:/Users/Hamid/Downloads/WeatherAUS.csv')
```

```python
df.describe()
```

```python
df = df.drop(['RainToday', 'RainTomorrow'], axis =1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

X = df.values
X = np.nan_to_num(X)
Clustering_df = StandardScaler().fit_transform(X)
Clustering_df
```

```python
# Elbow Method
sse = {}
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(X)
    sse[k] = kmeans.inertia_  # Sum of squared distances to closest cluster center

plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.show()
```

```python
clusterNum = 4
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
df["Cluster"] = labels
df.head(5)
```

```python
df.groupby('Cluster').mean()
```

```python
from sklearn.decomposition import PCA

# Step 1: Perform PCA to reduce data to two dimensions
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(Clustering_df)

# Choose the optimal number of clusters as determined previously
optimal_cluster_num = 4  # Replace this with your chosen number of clusters
k_means = KMeans(n_clusters=optimal_cluster_num, init='k-means++', n_init=12)
k_means.fit(Clustering_df)

# Step 2: Assign cluster labels
cluster_labels = k_means.labels_

# Step 3: Plot the clusters
plt.figure(figsize=(8, 6))
unique_labels = np.unique(cluster_labels)
for label in unique_labels:
    plt.scatter(principalComponents[cluster_labels == label, 0], principalComponents[cluster_labels == label, 1], label=f'Cluster {label}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D PCA Clustering Visualization')
plt.legend()
plt.grid(True)
plt.show()
```
