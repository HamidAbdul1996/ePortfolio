```python

# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data
data1 = 200 * randn(1000) + 50
data2 = data1 + (150 * randn(1000) + 50)

# calculate covariance matri
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```

Impact on Covariance:
Increasing first variable in data 1 increases covariance. Third variable does nothing. Increasing random array slightly increases.
On d2, increasing first variable make data way more disperesed but covariance actually slightly increases. No change on third variable.


Impact on Pearson Coefficient: D1, Increasing first varibale leads to big increase, makes the linear trend more pronounced. on d2, big decrease in increase of first variable.
